```python
import numpy as np
import matplotlib.pyplot as plt

import galpy
from galpy import potential
from galpy.orbit import Orbit
```

```python
v0=220 #km/s
r0=8e3 #pc
G_c=4.30091e-3 #pc (km/s)^2 Msun^-1
```

# 1.

```python
kepler = galpy.potential.KeplerPotential()
```

```python
o = Orbit([1, 0, 1.3, 0, 0, 0], ro=8., vo=220.)

times = np.linspace(0, 50, 1000)

o.integrate(t=times, pot=kepler)
```

```python
o.animate(d1='x', d2='y')
```

```python
kepler.plotEscapecurve(Rrange=[0.01,10.],grid=1001)
plt.grid()

plt.axvline(x=1, linestyle='--', color='k')
plt.axhline(y=1.4, linestyle='--', color='k')
```

Escape velocity for Kepler potential is about ~1.4$v_0$

# 2.

```python
isoch = galpy.potential.IsochronePotential()
```

```python
o = Orbit([1, 0, 0.4, 0, 0, 0], ro=8., vo=220.)

times = np.linspace(0, 50, 1000)

o.integrate(t=times, pot=isoch)
```

```python
o.animate(d1='x', d2='y')
```

```python
isoch.plotEscapecurve(Rrange=[0.01,10.],grid=1001)
plt.grid()

plt.axvline(x=1, linestyle='--', color='k')
plt.axhline(y=0.9, linestyle='--', color='k')
```

Escape velocity for isochrone potential is about ~0.9$v_0$
